### Stopwords NLTK

```python
# Stopwords bahasa Indonesia dari NLTK
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk_stopwords = set(nltk.corpus.stopwords.words('indonesian'))
```

### Stopwords CSV

```python
# Muat stopwords kustom dari CSV
stopwords_path = '/content/drive/MyDrive/TSA Dispertaru DS/Data Pasar DIY/stopwordbahasa.csv'
custom_stopwords = pd.read_csv(stopwords_path, header=None)
custom_stopwords = set(custom_stopwords[0].values)
```

### All Stopwords (Jika Butuh)

```python
all_stopwords = custom_stopwords.union(nltk_stopwords)
```

### Kamus Alay (Jika Butuh)

```python
slang_path = '/content/drive/MyDrive/TSA Dispertaru DS/Data Pasar DIY/kamusalay.csv'
slang_df = pd.read_csv(slang_path, header=None, encoding='ISO-8859-1')
slang_words = slang_df[0].tolist()  # Kolom pertama: kata slang
correct_words = slang_df[1].tolist()  # Kolom kedua: kata yang benar
slang_dict = dict(zip(slang_words, correct_words))
```

### Preprocessing Data

```python
# Preprocessing Data dalam satu fungsi
def preprocess_text(text, slang_dict, stopwords, stemmer):
    # Case folding
    text = text.lower()

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Tokenize
    tokens = word_tokenize(text)

    # Replace slang words
    tokens = [slang_dict.get(word, word) for word in tokens]

    # Remove stopwords
    tokens = [word for word in tokens if word not in stopwords]

    # Stemming
    tokens = [stemmer.stem(word) for word in tokens]

    # Gabungkan kembali token yang sudah di-preprocessing
    return ' '.join(tokens)
```

```python
# Gabungkan token yang telah di-stem ke dalam satu string
# df_pasar['processed_text'] = df_pasar['stemmed'].apply(lambda x: ' '.join(x))
df_pasar['processed_text'] = df_pasar['Ulasan'].apply(lambda text: preprocess_text(text, slang_dict, custom_stopwords, stemmer))
df_pasar.head()
```
