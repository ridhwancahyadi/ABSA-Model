```python
import pandas as pd
import nltk
import re
import emoji
from nltk.tokenize import word_tokenize
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

import nltk
nltk.download('punkt_tab')
```

## Data Cleansing

```python
df[df.isnull().any(axis=1)]
```

```python
df.duplicated().sum()
```

```python
def clean_text(text):
    # Menghapus URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Menghapus emoji
    text = emoji.replace_emoji(text, replace="")
    # # Mengganti tanda '- dan ,' dengan spasi
    text = re.sub(r'[-,]', ' ', text)
    # Menghapus semua simbol, termasuk koma
    text = re.sub(r'[^a-zA-Z\s ]', '', text)
    # Menghapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()
    # Mengubah ke huruf kecil
    text = text.lower()
    return text

# Membersihkan kolom ulasan
df['Cleansing'] = df['Ulasan'].apply(clean_text)
df
```

```python
df.isnull().sum()
```

## Tokenization

```python
def tokenize(text):
    tokens = word_tokenize(text)
    return tokens

# Membersihkan kolom ulasan
df['Tokens'] = df['Cleansing'].apply(tokenize)
df
```

## Normalize

```python
# Membaca kamus alay dan kamus singkatan
slang_path = '/content/drive/MyDrive/Tugas Akhir Skema Magang/Preprocessing/kamusalay.xlsx'
kamus_singkatan = pd.read_excel('/content/drive/MyDrive/Tugas Akhir Skema Magang/Preprocessing/Singkatan_Pasar_Beringharjo.xlsx')
slang_df = pd.read_excel(slang_path, header=None)
slang_dict = dict(zip(slang_df[0], slang_df[1]))  # Mengubah kamus alay menjadi dictionary
```

```python
# Fungsi normalisasi
def normalisasi(tokens, slang_dict, kamus_singkatan):
    # Ganti singkatan dengan makna yang benar
    tokens = [kamus_singkatan.loc[kamus_singkatan['Singkatan'] == token, 'Makna'].iloc[0]
              if token in kamus_singkatan['Singkatan'].values else token
              for token in tokens]
    # Ganti slang dengan kata yang benar
    return [slang_dict.get(word, word) for word in tokens]

# Terapkan normalisasi ke DataFrame
df['Normalize'] = df['Tokens'].apply(lambda tokens: normalisasi(tokens, slang_dict, kamus_singkatan))
df
```

```python
# Inisialisasi stopword remover
factory = StopWordRemoverFactory()
stopwords = factory.get_stop_words()

# Muat stopwords kustom dari CSV
stopwords_path = '/content/drive/MyDrive/TSA Dispertaru DS/Data Pasar DIY/stopwordbahasa.csv'
custom_stopwords = pd.read_csv(stopwords_path, header=None)
custom_stopwords = set(custom_stopwords[0].values)

# Menghapus stopwords menggunakan Sastrawi
def remove_stopwords(tokens):
    return [word for word in tokens if word not in stopwords]

df['Stopwords Removal'] = df['Normalize'].apply(remove_stopwords)
df
```

## Stemming

```python
# Inisialisasi stemmer Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi Stemming
def stemming(tokens):
    return [stemmer.stem(word) for word in tokens]

df['Stemming'] = df['Stopwords Removal'].apply(stemming)
df
```

```python
df['processed_text'] = df['Stemming'].apply(lambda x: ' '.join(x))
df.to_excel('pre dataset.xlsx', index=False)
```

```python
df.drop(columns=['Tokens', 'Normalize', 'Stopwords Removal', 'Stemming'], inplace=True)
df
```
